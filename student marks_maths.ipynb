```python
import pandas as pd 
# loading data 
df = pd.read_csv("desktop/student-mat.csv",delimiter = ';')
df.head()
```

```python
#extracting useful columns
df1 = df[['absences','G1','G2','G3','failures','studytime']]
df1.head()
```

```python
import numpy as np
#randomly splitting data sets into train and test
msk = np.random.rand(len(df)) < 0.8
train = df1[msk]
test = df1[~msk]
```

```python
from sklearn import linear_model
#applying linear regression algo
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['absences','G1','G2','failures','studytime']])
y = np.asanyarray(train['G3'])
regr.fit(x,y)

print ('Coefficients: ', regr.coef_)
print ('intercept: ', regr.intercept_)
```

```python
#accuracy
regr.score(test[['absences','G1','G2','failures','studytime']],test[['G3']])
```

```python
# to check for the best accuracy
a = []
for i in range(10):
    msk = np.random.rand(len(df)) < 0.8
    train = df1[msk]
    test = df1[~msk]
    
    
    regr = linear_model.LinearRegression()
    x = np.asanyarray(train[['absences','G1','G2','failures','studytime']])
    y = np.asanyarray(train['G3'])
    regr.fit(x,y)
    
    l1 = regr.score(test[['absences','G1','G2','failures','studytime']],test[['G3']])
    a.append(l1)
    
    
    
print(max(a))
    
    
    
    
        
    


        
    
    
    
```
