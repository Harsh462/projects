#### Converting financial distress columns into catrgorical variables

## Feature Engineering

### Balancing imbalance dataset

```python
dep_feat = fin_dis.iloc[:,-1]
ind_feat = fin_dis.iloc[:,0:85]
```

```python
ind_feat.head()
```

```python
!pip install imblearn
from imblearn.over_sampling import RandomOverSampler
```

```python
ros= RandomOverSampler(random_state = 0)
ind_bal ,dep_bal = ros.fit_resample(ind_feat,dep_feat)
```

```python
print(ind_bal.shape,dep_bal.shape)
```

```python
fin_dis = pd.concat([ind_bal,dep_bal],axis  = 1)
```

```python
fin_dis.head()
```

```python
top_10 = [ x for x in fin_dis["x80"].value_counts().sort_values(ascending  = False).head(10).index]
top_10
```

```python
def top_most(df,var,top_x):
    
    for digit in top_x:
            df[var+ "_" + str(digit)] = np.where(df[var] == digit,1,0)
            
```

```python
top_most(fin_dis,"x80",top_10)
```

```python
fin_dis.head()
```

```python
fin_dis.drop(columns = {"x80"},axis = 1,inplace = True )
```

## Feature selection

```python
dep_var = fin_dis.iloc[:,84]
ind_var1 = fin_dis.iloc[:,0:84]
ind_var2 = fin_dis.iloc[:, 85:]
ind_var = pd.concat([ind_var1,ind_var2],axis =1)
ind_var.head()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
etc = ExtraTreesClassifier()
etc.fit(ind_var,dep_var)
```

```python

best_feat = pd.Series(etc.feature_importances_,index = ind_var.columns)
best_feat
```

```python
best_feat.nlargest(15).plot(kind = "barh")
```

```python
best_feat.nlargest(15).index
```

#### These features are taken into consideration foe model building.

```python
model_df = fin_dis[['x2', 'x13', 'x3', 'x10', 'x5', 'x55', 'x14', 'x49', 'x24', 'x9', 'x53',
       'x23', 'x36', 'x50', 'x60']]
model_df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(model_df,dep_var,test_size = 0.3,random_state = 0)
```

### Random Forest classifier

```python
 
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 15)
```

```python
model = rfc.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import f1_score
```

```python
print(" F1 -score:" ,f1_score(y_test,pred))
```


### Support vector classifier

```python
from sklearn.svm import SVC
sv = SVC(C = 3,kernel = "linear")
```

```python
model1 = sv.fit(x_train,y_train)
```

```python
pred1 = model1.predict(x_test)
```

```python
pred1
```

```python
f1_score(y_test,pred1)
```
